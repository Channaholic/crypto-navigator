```python
import pandas as pd
import numpy as np
import requests
import time
```

```python
url = "https://api.coingecko.com/api/v3/"
headers = { 'Content-type': 'application/json'}
```

```python
r = requests.get(url + 'ping', headers = headers).json()
```

```python
r
```

```python
payload = {'vs_currency': 'usd', 'per_page':'100'}
```

Pull markets from Coingecko

```python
r = requests.get(url + 'coins/markets', params=payload, headers = headers).json()
# Work with Pagination for more than 250 coins
```

```python
df = pd.DataFrame(r)
```

```python
df.head()
```

```python
coin_data = []


for index, row in df.iterrows():
    time.sleep(5) #timer to circumwent API limits
    coin_id = row.id
    payload = {'coin_id': coin_id}
    coin_data.append(requests.get(url + 'coins/' + coin_id, params=payload, headers = headers).json())
    
```

```python
coin_data_df = pd.DataFrame(coin_data)
```

```python
final_cg = df.merge(coin_data_df, how='outer', on='id')
```

```python
final_cg = final_cg.add_suffix( "_cg")
final_cg.tail()
```

```python
final_cg.to_csv('../data/cg.csv')
```
