```python
import pandas as pd
import numpy as np
import requests
import json
```

```python
with open('../messari.txt', 'r') as file:
    API_KEY = file.read().replace('\n', '')
```

```python
url = "https://data.messari.io/api/v2/"
headers = { 'Content-type': 'application/json', "x-messari-api-key": API_KEY}
payload = {'page': '1', 'limit':'100'}
```

```python
r = requests.get(url + 'assets',params=payload, headers = headers).json()
```

```python
df = pd.DataFrame.from_dict(r['data'])
df = df.add_suffix( "_ms")
```

```python
df.to_csv('../data/messari.csv')
```
