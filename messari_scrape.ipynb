```python
import pandas as pd
import numpy as np
import requests
import json
```

```python
url = "https://data.messari.io/api/v2/"
headers = { 'Content-type': 'application/json'}
payload = {'page': '1'}
```

```python
r = requests.get(url + 'assets',params=payload, headers = headers).json()
```

```python
df = pd.DataFrame.from_dict(r['data'])
```
